## Algunos Conceptos 

### Grafos
Representan conjuntos de objetos interconectados entre ellos

### vertices o nodos 
Son los puntos o nodos con los que esta conformado un grafo. Es decir son objectos conectados o relacionados a otro objectos o No 

### Aristas o Bordes :
Son las líneas con las que se unen las aristas de un grafo y con la que se construyen también caminos

## Tipos de grafos.
Podemos clasificar los grafos en dos grupos: **dirigidos y no dirigidos.**

### Grafo no Dirigido
Hay nodos conectados unos con otros sin ninguna jerarquía u orden 

### Grafo Dirigido
Hay un orden de conexión y jerarquía donde un nodo puede seguir una dirección de conexión

```python
import networkx as nx
import random as rd
import matplotlib.pyplot as plt
from pyvis.network import Network
```

### Iniciar un grafo aleatorio no dirigidos

```python
#Iniciar un grafo aleatorio no dirigidos 
g_no_dir = nx.gaussian_random_partition_graph(n=25, s=4, v=5, p_in=0.25,p_out= 0.1,directed= False , seed=23) 

#Trazar la estructura del grafo
nx.draw(g_no_dir, with_labels= True )
```

### Visualización con pyvis

```python
nt = Network("400px", "500px", notebook=True)
nt.from_nx(g_no_dir)
nt.show('nx.html')
```

### Iniciar un grafo aleatorio dirigidos 

```python
#Iniciar un grafo aleatorio dirigidos 
g_dir= nx.gaussian_random_partition_graph(n=25, s=4, v=5, p_in=0.25,p_out= 0.1,directed= True , seed=23) 

#Trazar la estructura del grafo
nx.draw(g_dir, with_labels= True )
```

### Visualización con pyvis

```python
nt = Network("400px", "500px", notebook=True,directed =True)
nt.from_nx(g_dir)
nt.show('nx.html')
```

La única diferencia es directed=True. Tanto en networkx y pyvis. **Así de Fácil !!**

## Ponderado y No Ponderado

Otra propiedad que tienen los gráficos son las relaciones ponderadas entre nodos. Esto significa que la relación entre dos nodos se mide por un valor numérico que indica la magnitud de cuán fuerte, importante o costosa es la relación. De esta forma, los gráficos ponderados presentan pesos en todas las aristas, mientras que un gráfico no ponderado no (en este caso, todas las aristas tienen la misma importancia).

Creamos un gráfico con 35 nodos y 50 aristas aleatorias con la función gnm_random_graph, luego asignamos las propiedades 'pesos' a las aristas con una probabilidad aleatoria. Además, supongamos ahora que es una red social, donde cada nodo es una persona que tiene diferentes conexiones con amigos o familiares, si el peso es mayor a 0.6 entonces hay una conexión más cercana, entonces son familiares, pero si el peso es menos de 0,6 entonces la relación es menos fuerte y son amigos.

```python
#Inicializar un gráfico aleatorio 
G=nx.gnm_random_graph(35, 50, seed=12)
#Asigne pesos aleatorios 
for (u, v) in G.edges(): 
  # a cada borde se le agrega la propiedad ["weight"] que es en este caso aleatorio 
  G.edges[u,v]['weight'] = rd.random()

#Definir tipo de relación - La familia tiene más importancia que los amigos

familia = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] >= 0.6]
amigo   = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] < 0.6]
#Trama - Relación familiar (borde rojo) - Amistad (borde azul)
pos = nx.spring_layout(G, seed=124) 
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_edges(G, pos, edgelist=familia, width=2, edge_color='r', alpha = 0.4)
nx.draw_networkx_edges(G, pos, edgelist=amigo, width=1, edge_color='b', alpha= 0.4)
nx.draw_networkx_labels(G, pos)
ax = plt.gca() 
ax.margins(0.08) 
plt.axis("off") 
plt.tight_layout() 
plt.show()
```

**El atributo data=True** 

Los bordes se conectan con los nodos con dos valores (node1, node2) que establece la conexión, el atributo data=True devuelve una tupla con 3 atributos donde el tercero es la propiedad del borde (node1,node2,{'weight':0.74}). Si es Falso, devuelve 2 tuplas (u, v).

Aunque este es un ejemplo simple, en muchas aplicaciones del mundo real es importante tener en cuenta los bordes ponderados, ya que no todas las relaciones son iguales. 

## Gráficos dispersos y densos
**La densidad del gráfico se refiere tanto al número de nodos como a lo conectados que están.** En matemáticas, un gráfico denso es aquel en el que el número de aristas está cerca del número máximo de aristas. En cristiano esto quiere decir que si el numero máximo de conexiones es 15 entonces el numero de aristas promedio esta cerca del valor 15.


Si el gráfico es un grupo de islas de nodos desjuntas, podemos decir con seguridad que es un gráfico disperso.

### Ejemplo de uso 
para cambiar la densidad de un grafo se cambia la probabilidad. para poca densidad en este ejemplo sera de 0,08 y para el gráfico denso establecemos una probabilidad de 0,20.


```python
f, axs = plt.subplots(2,1,figsize=(15,15))

axs[0].set_title('Grafo poco denso')
g_dense = nx.erdos_renyi_graph(25, 0.08,seed=23)
nx.draw_networkx(g_dense, ax=axs[0], with_labels=True)
axs[1].set_title('Grafo muy Denso')
g_dense2 = nx.erdos_renyi_graph(25, 0.20,seed=23)
nx.draw_networkx(g_dense2, ax=axs[1], with_labels=True)
```

## Análisis de Centralidad
Existe una gran cantidad de medidas de centralidad para determinar la "importancia" relativa de un nodo en una red compleja. Esto determina la importancia de un nodo dentro de la red. 

Algunos de los grados de centralidad son.

### Centralidad de grados
La centralidad de grado es una de las medidas más simples de centralidad.
Mide el número de enlaces o conexiones que tiene un nodo con los demás nodos pertenecientes a un grafo. Si se tratara de un análisis de redes sociales implicaría grado de popularidad 


 

```python
d = nx.degree_centrality(g_dense2)
for v in g_dense2.nodes():
    print("%0.2d %5.3f" % (v, d[v]))# En este caso el nodo 0 es el que tiene mayor conexiones (popularidad)
```

### Centralidad por cercanía o proximidad

Mide la distancia media de un nodo a cualquier otro nodo del grafo. Por lo que los nodos que tengan la mayor puntuación tendrán, de media, distancias más cortas con el resto de los nodos. Esto significa que al tener más cercanía son más eficientes, rápidos en transmitir la información. Si habláramos de redes sociales la centralidad de grado sería un político con muchas conexiones y la centralidad por cercanía sería un periodista que comunica directamente con el público.





```python
c = nx.closeness_centrality(g_dense2)
for v in g_dense2.nodes():
    print("%0.2d %5.3f" % (v, c[v])) #En este caso el nodo cero es el que tiene mayor cercania con los demas nodos
```

### Centralidad de intermedicación
Esta es una medida de centralidad que cuantifica la frecuencia o el número de veces que un nodo actúa o sirve de puente dentro de una ruta corta entre dos nodos determinados.

Cuando en un grafo existen nodos de alta intermediación, estos suelen jugar un rol importante en la estructura a la que pertenecen. Estos nodos también poseen capacidades de ser controladores o reguladores de los flujos de información dentro de la estructura total del grafo. Esto es interesante de nuevo para aplicarlo en redes sociales. 


```python
b = nx.betweenness_centrality(g_dense2)
for v in g_dense2.nodes():
    print("%0.2d %5.3f" % (v, b[v]))
```
